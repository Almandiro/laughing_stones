```python
import pandas as pd
import json
import os
from pprint import pprint
```

```python
csv_list = list(filter(lambda x : x.endswith(".csv"), os.listdir("Data")))
csv_list
```

```python
dfs = []
for i, csv in enumerate(csv_list):
    print("Processing {0} of {1}  {2}".format(i + 1, len(csv_list), csv))
    df = pd.read_csv("Data/{}".format(csv))
    df["country"] = csv.split('.')[0]
    dfs.append(df)
```

```python
master_df = pd.concat(dfs)
master_df.head()
```

```python
del dfs
master_df.info()
```

```python
master_df.to_csv("Output/songs.csv")
```

```python
countries = None
with open("Data/country.json") as json_file:
    countries = pd.DataFrame.from_dict(json.load(json_file), orient="index", columns=["Name"]).reset_index()
countries.columns = ["Code", "Name"]
countries.to_csv("Output/countries.csv")
countries.head()
```
